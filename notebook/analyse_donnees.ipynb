```python
import scipy.stats as stats

# Effectuer le test ANOVA pour la variable 'carburant'
result = stats.f_oneway(data['prix'][data['carburant'] == 'essence'], data['prix'][data['carburant'] == 'diesel'])

print("Statistique F:", result.statistic)
print("Valeur p:", result.pvalue)
```

```python
import scipy.stats as stats

# Liste des variables catégorielles
categorical_columns = ['etat_de_route', 'carburant', 'turbo', 'nombre_portes', 'type_vehicule', 'roues_motrices', 'emplacement_moteur', 'marque', 'modele']

# Effectuer le test ANOVA pour chaque variable catégorielle
for column in categorical_columns:
    categories = data[column].unique()
    groups = [data['prix'][data[column] == category] for category in categories]
    result = stats.f_oneway(*groups)
    
    print(f"Variable : {column}")
    print("Statistique F:", result.statistic)
    print("Valeur p:", result.pvalue)
    print()
```
